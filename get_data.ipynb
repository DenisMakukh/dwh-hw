```python
import pandas as pd
```

```python
files = ['data/jan2022.csv', 'data/feb2022.csv', 'data/mar2022.csv', 'data/apr2022.csv']
```

```python
dataframes = [pd.read_csv(file) for file in files]
merged_data = pd.concat(dataframes, ignore_index=True)
```

```python
merged_data
```

```python
merged_data.to_csv('2022_jan_to_apr.csv', index=False)
print("Данные успешно объединены в файл 2022_jan_to_apr.csv!")
```
